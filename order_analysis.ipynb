```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub

# path = kagglehub.dataset_download("ankitbansal06/retail-orders")

order_data = pd.read_csv("orders.csv", na_values=["Not Available", "unknown"])
order_data.head(20)
```

```python
order_data["Ship Mode"].unique() #get distinct values
```

```python
# renaming columns as they have spaces in between, and lowercasing them
order_data.columns = order_data.columns.str.lower().str.replace(' ', '_')
```

```python
# make new columsn like discount, selling price, and net profit

order_data['discount'] = order_data['list_price'] * order_data['discount_percent'] * 0.01
order_data['selling_price'] = order_data['list_price'] - order_data['discount']
order_data['net_profit'] = order_data['selling_price'] - order_data['cost_price']
```

```python
#convert order date to datetime
import pandas as pd
order_data['order_date'] = pd.to_datetime(order_data['order_date'], format='%Y-%m-%d')
```

```python
order_data.drop(columns=['list_price', 'discount_percent', 'cost_price'], inplace=True)
```

```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.orm import declarative_base, sessionmaker

# Create SQLite database file
engine = create_engine("sqlite:///data_analytics_db.db", connect_args={"check_same_thread": False})  # This creates a file named 'my_database.db'

# Declare your table structure
Base = declarative_base()

class User(Base):
    __tablename__ = 'users'

    id = Column(Integer, primary_key=True)
    name = Column(String)
    email = Column(String)

# Create the tables
Base.metadata.create_all(engine)

# Create a session and insert data
Session = sessionmaker(bind=engine)
session = Session()

new_user = User(name="Diggy", email="diggy@data.com")
session.add(new_user)
session.commit()

# Querying
for user in session.query(User).all():
    print(user.name, user.email)
```

```python
connection = engine.connect()
```

```python
order_data.to_sql('order_data', engine, if_exists='append', index=False)
```
